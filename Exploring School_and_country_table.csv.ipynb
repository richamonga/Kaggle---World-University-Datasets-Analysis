## Data Cleansing/Data Preparation Steps

#### Feature='world_rank'

'world_rank' is our target or the attribute which depends on all other features on this data set.In order to be able to see relation between the dependent variable and feature variables,the prior should be in numeric format.However,if we notice clearly.After world_rank of 200,all the universities have been ranked in a range.Example,'world_rank' for Cardiff University in 2012 is '201-225'.This fact is true for 'world_rank' post '200' for every year.
Due to this finding correlation between 'world_rank' (which is our dependant variable) and other feature variables is not feasible.
Possible solutions : 
1. We can keep records with 'world_rank'<=200 in train data set and rest in test data set in order to predict suitable world rank for these records.
2. We can replace the records with rank in range with sequence valued ranks for further analysis on correlation.Example,if there are 25 records with 'world_rank' ='201-225',the we can assign each record a rank starting 201 till 225.

#### Feature ='country'

We have a issue of dirty data present in this column which can result in incorrect data analysis and reporting.

Example,row 2021 for University of Hawai is having country name as  <font ='3' color="blue"> 'Unisted States of America' </font>  instead of
United States of America.
We need to correct this issue of incorrect spelling of text data.

```python
df['country']=df['country'].str.replace("Unisted","United")
```

#### Feature = 'international'
This feature or attribute has not only numeric data stored in non-numeric format but its has also got presence of '-' character where ever data is not present.Therefore,we need to cleanse this data by replacing or imputing '-' by average value for this attribute across years for a particular university.We will further replace Null or NaN with 0 for those universities which have '-' as value across years.

Once values for this attribute are present in numeric format,it can be used to find correlation with Target and for other kind of data analysis.

```python
df['international']=pd.to_numeric(df['international'].map(str).apply(lambda x : x.strip('-')))
df['international']=df['international'].fillna(df.groupby(['university_name'])['international'].transform('mean'))
df['international']=df['international'].fillna(0)
```

#### Feature = 'income'
Like,'international',this feature also has same issues.It has numeric data stored in non-numeric format and has also got presence of '-' character where ever data is not present.Therefore,we need to cleanse this data by replacing '-' by average value for this attribute across years for a particular university.We will further replace leftover NaNs with 0 for those universities which have '-' as value across years.

```python
df['income']=pd.to_numeric(df['income'].map(str).apply(lambda x : x.strip('-')))
df['income']=df['income'].fillna(df.groupby(['university_name'])['income'].transform('mean'))
df['income']=df['income'].fillna(0)
df.income[df['university_name']=='University of Toronto']
```

#### Feature = 'total_score'
Like,'international' and 'income',this feature also has same issues.We will apply same data cleansing methods here too.As we did for features above,universities which are still left with Null or NaNs even after imputing with mean value,we will replace NaNs with 0 for such cases.

```python
df['total_score']=pd.to_numeric(df['total_score'].map(str).apply(lambda x : x.strip('-')))
df['total_score']=df['total_score'].fillna(df.groupby(['university_name'])['total_score'].transform('mean'))
df['total_score']=df['total_score'].fillna(0)
df.total_score[df['university_name']=='University of Kansas']
```

#### Feature = 'num_students'
This fields has to be converted to a numeric data type and ',' needs to removed to convert data into appropriate format.We also have NULLS/NaNs present in this attribute.We will prep this feauture to have mean value of num_students by grouping on university across years.We will still have 0 for those universities which donot have value for total_score available across years.

```python
df['num_students']=df['num_students'].str.replace(",","").astype(float) 
df['num_students']=df['num_students'].fillna(df.groupby(['university_name'])['num_students'].transform('mean'))
df['num_students']=df['num_students'].fillna(0)
```

#### Feature ='student_staff_ratio'

For this feature we have to handle issue of missing values by imputing with mean or replacing with 0 where data is not availale for a university across years.

```python
df['student_staff_ratio']=df['student_staff_ratio'].fillna(df.groupby(['university_name'])['student_staff_ratio'].transform('mean'))
df['student_staff_ratio']=df['student_staff_ratio'].fillna(0)
```

#### Feature = 'international_students'

To prepare this field for data analysis,we need to handle missing values along with removing '%' sign in order to convert this field to numeric data type which will enable us to use this feature for data analysis.

```python
df['international_students']=df['international_students'].str.replace("%","").astype(float) 
df['international_students']=df['international_students'].fillna(df.groupby(['university_name'])['international_students'].transform('mean'))
df['international_students']=df['international_students'].fillna(0)
df.international_students[df['university_name']=='University of Chicago']
```

```python
def answer_three():    
    df = pd.read_csv("timesData.csv")    
    df_subset=df[df['country'].isin(answer_two())].sort_values(['country', 'year'], ascending=[True, True])
    df_subset['international_students']=pd.to_numeric(df_subset['international_students'].map(str).apply(lambda x : x.strip('%')))
    x=df_subset.groupby('university_name')['international_students'].mean().nlargest(1)
    most_foreigner_friendly_universtity=x.index[0]
    return most_foreigner_friendly_universtity
answer_three()
```
